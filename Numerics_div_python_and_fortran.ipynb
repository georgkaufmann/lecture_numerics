## Simple code in `python`

```python
"""
**python_variables**
Defining variables and types in python
Georg Kaufmann
"""
#----------------------------------------------------------------------
# define integer, integer list, real, real list, string
#----------------------------------------------------------------------
i=1
j=[1,2,3,4]
x=1.0
y=[1.,2.,3.,4.]
z="Georg Kaufmann"
b=True
#----------------------------------------------------------------------
# print variable and its type
#----------------------------------------------------------------------
print ("-------------------------")
print ("variables and their types")
print ("-------------------------")
print ('i=%22s %s' % (i,type(i)))
print ('j=%22s %s' % (j,type(j)))
print ('x=%22s %s' % (x,type(x)))
print ('y=%22s %s' % (y,type(y)))
print ('z=%22s %s' % (z,type(z)))
print ('b=%22s %s' % (b,type(b)))
#----------------------------------------------------------------------
# merge all variables into new list
#----------------------------------------------------------------------
L=[i,j,x,y,z,b]
#----------------------------------------------------------------------
# examine list
#----------------------------------------------------------------------
string_len=len(L)
string_elements=range(string_len)
print ("-------------------------")
print ("variables merged into list")
print ("-------------------------")
for i in string_elements:
    print ('element L[%s] has a value of %22s and is of type %20s' % (i,L[i],type(L[i])))
```

## Simple code in `Fortran`

You need to compile the *source code* below, e.g. by saving it to a file
`fortran_variables.f90`, then use a compiler (e.g. gfortran)

~~~bash
$ gfortran fortran_variables.f90 -o fortran_variables
~~~
This compilation creates an *executable code* `fortran_variables`, which you can run
in a shell:
~~~bash
$ ./fortran_variables
~~~

```python
      program fortran_variables
!----------------------------------------------------------------------
! Defining variables and types in fortran
! Georg Kaufmann
!----------------------------------------------------------------------
      implicit none
      integer                             :: i,k
      integer, dimension(4)               :: j
      real*4                              :: x
      real*4, dimension(4)                :: y
      character*20                        :: z
      logical                             :: b
!----------------------------------------------------------------------
! assigning values
!----------------------------------------------------------------------
      i = 1
      j(1) = (1); j(2) = (2); j(3) = (3); j(4) = (4);
      x = 1.e0
      y(1) = (1); y(2) = (2); y(3) = (3); y(4) = (4);
      z = "Georg Kaufmann"
      b = .true.
!----------------------------------------------------------------------
! write to screen
!----------------------------------------------------------------------
      write (6,101) i,(j(k),k=1,4),x,(y(k),k=1,4),z,b

      stop
 101  format ('-------------------------',/, &
              'variables and their types',/, &
              '-------------------------',/, &
              'i=      ',i5,/, &
              'j(4)=   ',4i5,/, &
              'x=      ',f5.2,/, &
              'y(4)=   ',4f5.2,/, &
              'z=      ',a20,/, &
              'b=      ',l2)
      end program fortran_variables
```

## And in `fortran`

```python
      program fortran_intro
!----------------------------------------------------------------------
! declaration 
!----------------------------------------------------------------------
      implicit none
      integer                           :: n,a,b,tmp,x,i
      real*4                            :: c
      integer,dimension(:),allocatable  :: list1,list2,list,fib
      character(len=15)                 :: c1,c2
      character(len=7),dimension(4)     :: text
      integer                           :: fib2
      external                          :: fib2
!----------------------------------------------------------------------
! define some variables
!----------------------------------------------------------------------
      n = 1                    ! first variable, integer
      a = 2                    ! second variable, integer
      c = 2.                   ! third variable, float
      c1 = "Georg Kaufmann"    ! string, double quotes
      c2 = 'Georg Kaufmann'    ! string, single quotes
      write (6,*) n,a,c
      write (6,100) n,a,c
!----------------------------------------------------------------------
! lists
!----------------------------------------------------------------------
      allocate (list1(6),list2(2),list(8))
      list1(1:5) = (/ 1,2,4,8,16 /)
      list2(1:2) = (/ 2**6,2**7 /)
      write (6,*) list1,list2
      list1(6)   = 32
      list(1:8)  = (/ list1(1:6), list2(1:2)/)
      write (6,*) list
!----------------------------------------------------------------------
! do while loop
!----------------------------------------------------------------------
      a = 0; b = 1
      do while (a < 10)
        write (6,*) a
        tmp = a
        a = b; b = tmp + b
      enddo
!----------------------------------------------------------------------
! if
!----------------------------------------------------------------------
      write (6,101)
      read (5,*) x
      if ( x < 0) then
         write (6,*) 'x < 0'
      else if ( x == 0) then
         write (6,*) 'x = 0'
      else
         write (6,*) 'x > 0'
      endif
!----------------------------------------------------------------------
! do loop
!----------------------------------------------------------------------
      text = (/ 'Wir    ', 'werden ' , 'alle   ', 'sterben' /)
      do i = 1,4
        write (6,*) i,text(i)
      enddo
!----------------------------------------------------------------------
! subroutine
!----------------------------------------------------------------------
      call fib1(10)
!----------------------------------------------------------------------
! function
!----------------------------------------------------------------------
      allocate (fib(10))
      do i = 1,10
        fib(i) = fib2(i)
      enddo
      write (6,'(10i10)') (fib(i),i=1,10)
      deallocate (list1,list2,list,fib)
      stop
 100  format ("i = ",i5," a = ",i10," b = ",f10.2)
 101  format ('Enter x: ',$)
      end program fortran_intro


      subroutine fib1(n)
!----------------------------------------------------------------------
! Calculate Fibonacci numbers and print to screen
!----------------------------------------------------------------------
      implicit none
      integer                           :: n,a,b,tmp
      a = 0; b = 1
      do while (a < n)
        write (6,*) a
        tmp = a
        a = b; b = tmp + b
      enddo
      return
      end subroutine fib1
        

      function fib2(n)
!----------------------------------------------------------------------
! Calculate Fibonacci numbers and print to screen
!----------------------------------------------------------------------
      implicit none
      integer                           :: i,n,a,b,fib2
      a = 0; b = 1
      do i = 1,n
        fib2 = a
        a = b; b = fib2 + b
      enddo
      return
      end function fib2
```
