# Initial-value problems: Library version

```python
import numpy as np
import matplotlib.pyplot as plt
import csv

import numerics.initialvalue
print(dir(numerics.initialvalue))
import numerics.change_fonts;numerics.change_fonts.change_fontsize()
```

Initial-value problem:
$$
\begin{array}{rcl}
{{dy}\over{dt}} &=& y - t^2 + 1 \\
y(t=0) & = & \alpha
\end{array}
$$

Exact solution:
$$
y(t) = (t+1)^2 - 0.5 e^{t}
$$

```python
def rhs(t,y):
    '''
    #----------------------------------------------------------------------
    # right-hand side function rhs(x)
    #----------------------------------------------------------------------
    '''
    y = y - t**2 + 1
    return y

def fexact(t):
    # analytical solution
    y = (t + 1)**2 - 0.5*np.exp(t)
    return y
```

## Initial conditions

```python
a     = 0.
b     = 5.
alpha = 0.5
n = 21
h = (b-a) / float(n-1)
print ("%3s %5.2f %3s %5.2f %7s %5.2f %3s %4i %3s %5.2f" % ('a:',a,'b:',b,'alpha:',alpha,'n:',n,'h:',h))
```

## Euler method

```python

t = []; y1 = []; yexact = []
# initial values
t.append(a)
y1.append(alpha)
yexact.append(alpha)
print ("%4i %10.4f %10.2f %10.2f %10.2f" % (0,h,t[0],y1[0],yexact[0]))
# run over interal
for i in np.arange(1,n):
    t.append(t[i-1]+h)
    ynew = numerics.initialvalue.ivp_euler(rhs,t[i-1],y1[i-1],h)
    y1.append(ynew)
    yexact.append(fexact(t[i]))
    print ("%4i %10.4f %10.2f %10.2f %10.2f" % (i,h,t[i],y1[i],yexact[i]))
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('t')
plt.ylabel('y')
plt.plot(t,yexact,linewidth=4,color='gray',label='y(t)=(t+1)$^2$-0.5*e$^t$')
plt.plot(t,y1,linewidth=2,color='red',label='Euler method')
plt.legend()
```

## Modified Euler method

```python

t = []; y1 = []; y2 = []; yexact = []
# initial values
t.append(a)
y1.append(alpha)
y2.append(alpha)
yexact.append(alpha)
print ("%4i %10.4f %10.4f %10.2f %10.2f %10.2f" % (0,h,t[0],y1[0],y2[0],yexact[0]))
# run over interal
for i in np.arange(1,n):
    t.append(t[i-1]+h)
    y1.append(numerics.initialvalue.ivp_euler(rhs,t[i-1],y1[i-1],h))
    y2.append(numerics.initialvalue.ivp_eulermod(rhs,t[i-1],y2[i-1],h))
    yexact.append(fexact(t[i]))
    print ("%4i %10.4f %10.4f %10.2f %10.2f %10.2f" % (i,h,t[i],y1[i],y2[i],yexact[i]))
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('t')
plt.ylabel('y')
plt.plot(t,yexact,linewidth=4,color='gray',label='y(t)=(t+1)$^2$-0.5*e$^t$')
plt.plot(t,y1,linewidth=2,color='red',label='Euler method')
plt.plot(t,y2,linewidth=2,color='green',label='Mod. Euler method')
plt.legend()
```

## Runge-Kutta method

```python

t = []; y1 = []; y2 = []; y3 = []; yexact = []
# initial values
t.append(a)
y1.append(alpha)
y2.append(alpha)
y3.append(alpha)
yexact.append(alpha)
print ("%4i %10.4f %10.4f %10.2f %10.4f %10.4f %10.4f" % (0,h,t[0],y1[0],y2[0],y3[0],yexact[0]))
# run over interal
for i in np.arange(1,n):
    t.append(t[i-1]+h)
    y1.append(numerics.initialvalue.ivp_euler(rhs,t[i-1],y1[i-1],h))
    y2.append(numerics.initialvalue.ivp_eulermod(rhs,t[i-1],y2[i-1],h))
    y3.append(numerics.initialvalue.ivp_rk4(rhs,t[i-1],y3[i-1],h))
    yexact.append(fexact(t[i]))
    print ("%4i %10.4f %10.4f %10.2f %10.4f %10.4f %10.4f" % (i,h,t[i],y1[i],y2[i],y3[i],yexact[i]))
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('t')
plt.ylabel('y')
plt.plot(t,yexact,linewidth=4,color='gray',label='y(t)=(t+1)$^2$-0.5*e$^t$')
plt.plot(t,y1,linewidth=2,color='red',label='Euler method')
plt.plot(t,y2,linewidth=2,color='green',label='Mod. Euler method')
plt.plot(t,y3,linewidth=2,color='blue',label='Runge-Kutta method')
plt.legend()
```
