## Analytical solution of 1D heat-diffusion equation

```python
import numpy as np
import matplotlib.pyplot as plt
```

We seek solution for the **1D transient heat equation** for the field $c(t,x)$:
$$
\frac{\partial c}{\partial t} =K \frac{\partial^2 c}{\partial x^2}
$$
with
- $t$ [s] time
- $x$ [m] coordinate
- $K$ [m/s] thermal conductivity

for the **initial condition**
$$
c(0,x) = c_0(x)
$$
and the **boundary conditions**:
$$
\begin{array}{rcl}
 c(t,0) &=& 0 \\
 c(t,1) &=& 0
\end{array}
$$

Solutions are:
$$
\begin{array}{rcl}
c_0(x) &=& \sin(n \pi x) \\
c(t,x) &=& e^{-n^2 \pi^2 K t} \sin(n \pi x) 
\end{array}
$$
for $n=1,\dots$

```python
x = np.linspace(0,1,101)
A1 = 1.
K = 0.1
n = 1

def c0(x,n=1):
    # define initial condition
    c0 = np.sin(n*np.pi*x)
    return c0

def c(t,x,A,K,n=1):
    # define analytical solution
    c = A * np.exp(-(n*np.pi)**2 * K * t) * np.sin(n*np.pi*x)
    return c
```

```python
n=1
plt.figure(figsize=(10,5))
plt.title('1D heat-diffusion (n='+str(n)+')')
plt.xlabel('x [m]')
plt.ylabel('c')
plt.plot(x,c0(x,n=n),linewidth=5,color='black',alpha=0.5,label='c0')
t = [0.2,0.4,0.6,0.8,1.]
for i in range(len(t)):
    plt.plot(x,c(t[i],x,A1,K,n=n),label='t='+str(round(t[i],2))+' s')
plt.legend()
```

```python
n=2
plt.figure(figsize=(10,5))
plt.title('1D heat-diffusion (n='+str(n)+')')
plt.xlabel('x [m]')
plt.ylabel('c')
plt.plot(x,c0(x,n=n),linewidth=5,color='black',alpha=0.5,label='c0')
t = [0.2,0.4,0.6,0.8,1.]
for i in range(len(t)):
    plt.plot(x,c(t[i],x,A1,K,n=n),label='t='+str(round(t[i],2))+' s')
plt.legend()
```
