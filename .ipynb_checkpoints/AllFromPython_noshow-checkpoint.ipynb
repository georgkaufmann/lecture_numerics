# Numerical methods in geosciences in a nutshell
----

In this notebook, we will introduce all methods discussed in the lecture
`Numerical methods in geosciences` with `python` routines and functions,
based on a simple problem.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import scipy.optimize
import scipy.misc
import scipy.integrate
import scipy.interpolate
import numpy.polynomial
```

## Function

We will use a simple **sine** as function, for which we know the derivative(s).

Define the function
$$f(x)=\sin(x)$$
in the interval $x \in [a,b]$.

We use `x` and `y` as arrays for the **original** data set, which can be used for comparison later.

```python
a = -0.1
b = 2.*np.pi+0.1
x = np.linspace(a,b,101)

def f(x):
    # function f(x)=sin(x)
    f = np.sin(x)
    return f

y = f(x)

plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='$f(x)=\sin(x)$')
```

## Roots

The sine function has roots for $x^0=i \pi, i=-n,n$.

We search for the roots by bracketing them first ($f(x_i)f(x_{i+1})<0$), the use
the `newton` function from `scipy.optimize`. This routine uses different
methods for the root finding:

- If we **do not** hand down a derivative of the function `fprime`, then the
**secant method** is used!
- If we hand down a derivative via the `fprime` flag, the **Newton method** is used.

```python
roots_secant = np.array([])
roots_newton = np.array([])
for i in range(len(x)-1):
    if (f(x[i])==0.):
        roots_secant = np.append(roots_secant,x[i])
        roots_newton = np.append(roots_newton,x[i])
    elif (f(x[i+1])==0.):
        roots_secant = np.append(roots_secant,x[i+1])
        roots_newton = np.append(roots_newton,x[i+1])
    elif (f(x[i])*f(x[i+1])<0):
        root=sp.optimize.newton(f,(x[i]+x[i+1])/2)
        roots_secant = np.append(roots_secant,root)
        root=sp.optimize.newton(f,(x[i]+x[i+1])/2,fprime=df)
        roots_newton = np.append(roots_newton,root)
    
print(roots_secant)
print(roots_newton)
plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='$f(x)=\sin(x)$')
plt.plot(roots_newton,f(roots_newton),linewidth=0,marker='o',markersize=15,color='red',label='Newton')
plt.plot(roots_secant,f(roots_secant),linewidth=0,marker='x',markersize=12,color='blue',label='Secant')
plt.legend()
```

## Derivatives

We calculate the first and second derivative of the sine function:
$$
\begin{array}{rcl}
f'(x)  &=& \frac{df(x)}{dx} = \frac{d}{dx} \sin(x) = \cos(x) \\
f''(x) &=& \frac{d^2f(x)}{dx^2} = \frac{d^2}{dx^2} \sin(x) = -\sin(x)
\end{array}
$$
using the `derivative` function of the `scipy.misc` package.

We hand down the $n$-th derivative ($n=1$ and $n=2$ in our case) and a tolerance `dx`.
Note that the tolerance `dx` is preset as `dx=1`, thus when you do not define
this parameter by yourself, weird results might occur...

```python
def df(x):
    df = scipy.misc.derivative(f,x,n=1,dx=1.e-5)
    return df

def d2f(x):
    d2f = scipy.misc.derivative(f,x,n=2,dx=1.e-5)
    return d2f

plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='f')
plt.plot(x,df(x),linewidth=3,color='red',label="f$^{\'}=\\frac{df}{dx}$")
plt.plot(x,d2f(x),linewidth=3,color='green',label='f$^{\'\'}=\\frac{d^2f}{dx^2}$')
plt.legend()
```

## Integration

We integrate the sine function over the interval [a,b]:
$$
I = \int\limits_a^b sin(x)dx = [-cos(x)]^b_a = -cos(b) - (-cos(a))
$$
Should result in $I=0$ for $[0,2\pi]$.

We use two methods:

- the **simpson rule** `simps` from the `scipy.integrate` package,
- the **quadrature**, also from the `scipy.integrate` package.

Note that in ` scipy.integrate.simps`, the function is explicitely handed down as array!

```python
Isimpson = scipy.integrate.simps(f(x), x)
print ('Simpson:    ',Isimpson)

Igauss = scipy.integrate.quad(f,a,b)
print ('Quadrature: ',Igauss[0])

plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.fill_between(x,f(x),0.)
```

## ODE

We rewrite the derivative of the sine function as **ordinary differential equation** of first order (ODE):
$$
\begin{array}{rcl}
\frac{df}{dx}=y' &=& cos(x)\\
y(x=a) &=& 0
\end{array}
$$
This is a **first-order initial value problem** with the initial
condition $y(a)=f(x)$.

We solve this ODE with the `odeint` solver of the `scipy.integrate` package.
Note that the `args` keyword is used as dummy here to show, how additional
arguments for the right-hand side function can be passed down.

```python
# define the right-hand side
def rhs(y,x,args=(a)):
    dydt = np.cos(x)
    return dydt

# define the initial condition
y0 = [np.sin(a)]
print(y0)

# solve the ODE for the given range
sol1 = scipy.integrate.odeint(rhs, y0, x, args=(a,))

plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='$f(x)=\sin(x)$')
plt.plot(x,sol1,linewidth='2',color=(1,0,0),label='odeint')
plt.legend()
```

We solve this ODE with the `solve_ivp` solver of the `scipy.integrate` package.

Note that for this package, the order of $x$ and $y$ for the right-hand side function are reversed!

```python
# define the right-hand side
def rhs2(x,y,args=(a)):
    dydt = np.cos(x)
    return dydt

# define the initial condition
y0 = [np.sin(a)]
print(y0)

# solve the ODE for the given range
sol2 = scipy.integrate.solve_ivp(rhs2,[x[0],x[-1]],y0,t_eval=x,args=(a,))
#print(sol2)

plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='$f(x)=\sin(x)$')
plt.plot(x,sol1,linewidth='2',color=(1,0,0),label='odeint')
plt.plot(x,sol2.y[0],linewidth='1',linestyle='dashed',color=(0,0,0),label='solve_ivp')
plt.legend()
```

## Interpolate

We sample the sine function with $n$ points, then use a **Lagrange interpolation polynomial** to
approximate the sampled function.

From the `scipy.interpolate` package, we call the `lagrange` function, which returns the $n-1$ coefficients $a$
of the Lagrange interpolation polynomial:
$$
P_{n-1}(x) = \sum\limits_k L_{n,k}(x) f(x_k) = \sum\limits_n a_n x^n
$$

```python
# sampling
n  = 8
xx = np.linspace(a,b,n)
yy = f(xx)

# Lagrange polynomial
poly = scipy.interpolate.lagrange(xx, yy)
coeff=numpy.polynomial.Polynomial(poly).coef
print(coeff)

# re-create polynomial
def polynom(x,coeff):
    n=len(coeff)
    polynom = 0.
    for i in range(n):
        polynom = polynom + coeff[i]*x**(n-1-i)
    return polynom

plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='$f(x)=\sin(x)$')
plt.plot(xx,yy,linewidth=0,marker='o',markersize=15,color='blue',label='Samples')
plt.plot(xx,polynom(xx,coeff),linewidth=2,linestyle='--',color='red',label='Lagrange')
plt.legend()
```

We add a **cubic spline** from the `scipy.interpolate` package.

```python
# cubic spline
spline = scipy.interpolate.CubicSpline(xx,yy)

plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='$f(x)=\sin(x)$')
plt.plot(xx,yy,linewidth=0,marker='o',markersize=15,color='blue',label='Samples')
plt.plot(x,spline(x),linewidth=2,color='lightgreen',label='Cubic spline')
plt.plot(xx,polynom(xx,coeff),linewidth=2,linestyle='--',color='red',label='Lagrange')
plt.legend()
```

## Fitting

We fit a self-defined function through a set of $n$ sample points $x_i,i=1,n$.
$$
fit(x,a,b) = a \sin(bx)
$$
If the use as data our sinus, $f(x)=\sin(x)$, and define $n$ sample points $x_i,i=1,n$,
the **two fitting parameters** $a$ and $b$ should be:
$$
\begin{array}{rcl}
a&=&1\\
b&=&1
\end{array}
$$
Once we add noise to our data, $f(x)=\sin(x)+noise(x)$, the fit will become worse.

We use the `fit` algorithm from the `scipy.optimize` package.

- We first create $n$ data points with $x$ and $y$ values, and add some noise.
- We then define a fit function `fit`, used for the non-linear fitting routine.
- We then call `scipy.optimize.curve_fit`. Upson succes, the optimized fitting
parameter are returned in `params`.

```python
dy = 0.15
n  = 8
xx=np.linspace(a,b,n)
yy=f(xx)
yy=yy+np.random.uniform(-dy,dy,len(xx))
sigma=[dy for i in range(len(xx))]

def fit(x, a, b):
    return a * f(b*x)

params, params_covariance = scipy.optimize.curve_fit(fit, xx, yy, sigma=sigma)
print('a,b:   ',params)
print('covar: ',params_covariance)


plt.figure(figsize=(10,6))
plt.xlim([a,b])
plt.ylim([-1.1,1.1])
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x)=\sin(x)$')
plt.plot(x,y,linewidth=8,color='gray',label='$f(x)=\sin(x)$')
plt.plot(xx, yy, marker='o',markersize=10,linewidth=0,label='Data',color='blue')
plt.plot(xx, fit(xx, params[0], params[1]),linewidth=2,color='red',label='Fitted function')
plt.plot(xx, fit(xx, *params),linewidth=2,color='red',label='Fitted function')
plt.legend(loc='best')
```
