```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2D Laplace equation

We seek solutions of the Laplace equation for the 2D field $c(x,y)$:
$$
 \frac{\partial^2 c}{\partial x^2} + \frac{\partial^2 c}{\partial y^2} = 0
$$
- $x$ [m] coordinate
- $y$ [m] coordinate

for a set of boundary conditions:
$$
\begin{array}{rcl}
 c(0,y) &=& 0 \\ 
 c(x,0) &=& 0 \\
 c(a,y) &=& \sinh(x) \\
 c(x,b) &=& \sin(y)
\end{array}
$$

Solutions are:
$$
c(x,y) = \sinh(\lambda x) \sin(\lambda y)
$$
with $\lambda$ an eigenvalue.

```python
x = np.linspace(-3,3,101)
plt.plot(x,np.sinh(x),label='sinh')
plt.plot(x,np.cosh(x),label='cosh')
plt.legend()
plt.grid()
```

```python
a = np.pi
b = 3*np.pi

nx=100;ny=50
x=np.linspace(0,a,nx)
y=np.linspace(0,b,ny)
```

```python
upper = np.sinh(x)
right = np.sin(y)
```

```python
%matplotlib inline
fig,ax = plt.subplots(1,2)
ax[0].plot(x,upper)
ax[1].plot(y,right)
```

```python
lamb=1
X,Y = np.meshgrid(x,y)
U = np.sinh(lamb*X) * np.sin(lamb*Y)
print(U.shape,U.max())
U=U/U.max()
```

```python
%matplotlib inline
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')
ax.set_title('Laplace equation ($\lambda=$'+str(lamb)+')')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('c')
ax.plot_surface(X,Y,U)
```

```python
lamb=2
X,Y = np.meshgrid(x,y)
U = np.sinh(lamb*X) * np.sin(lamb*Y)
print(U.shape,U.max())
U=U/U.max()
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')
ax.set_title('Laplace equation ($\lambda=$'+str(lamb)+')')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('c')
ax.plot_surface(X,Y,U)
```
