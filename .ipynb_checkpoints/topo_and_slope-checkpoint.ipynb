```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative


from numerics.change_fonts import *
change_fontsize()
```

```python
def f(x):
    y=x*np.exp(-np.abs(x)/4.)*500.
    #y=np.sin(x)*500.
    return y

xorig=np.linspace(-10,50,101)

np.random.seed(12)
xsample=np.sort(np.random.randint(-10,50,31))
print(xsample)

slope=derivative(f,xsample,n=1,order=3,dx=1.)
```

```python
plt.figure(figsize=(10,4))
plt.xlim([-12,52])
plt.ylim([-1000,1000])
plt.xlabel('Profile [km]')
plt.ylabel('Elevation [m]')

plt.fill_between(xorig,f(xorig),y2=-1000,linewidth=1,color='brown')
plt.scatter(xsample,f(xsample),marker='o',color='black',s=80)
```

```python
plt.figure(figsize=(10,4))
plt.xlim([-12,52])
plt.ylim([-1000,1000])
plt.xlabel('Profile [km]')
plt.ylabel('Slope [-]')
plt.fill_between(xorig,f(xorig),y2=-1000,linewidth=1,color='brown')
plt.plot(xsample,slope,linewidth=4,color='green')
```
